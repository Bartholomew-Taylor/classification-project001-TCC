# Assumptions (Questions): 

-There is a correlation between month-to-month and churn

-Customers that start their tenure with a high monthly cost are more likely to churn

-Fiber optics customers are more likely to churn

-there is a correlation between churn and senior citizen status


### -There is a dependence on month-to-month and churn

    null: there is no dependence
    
    alt: there is a dependence

```python
alpha = 0.05

```

```python
observed = pd.crosstab(telco_clean_df.contract_type, telco_clean_df.churn)
observed
```

```python
chi2, p, degf, expected = stats.chi2_contingency(observed)
chi2, p, degf, expected
```

#### p is less than alpha so the alt hypothesis is supported

## There is a correlation between fiber optics customers and churn

    Null: there is no correlation
    Alt: there is a correlation


```python
observed2 = pd.crosstab(telco_clean_df.internet_service_type, telco_clean_df.churn)
observed2
```

```python
chi2, p, degf, expected = stats.chi2_contingency(observed2)
chi2, p, degf, expected
```

### p is less than alpha so the alt hypothesis is supported

## There is a correlation betweeen Total Charge and Churn (with the assumption of tenure)
    Null: there is no correlation
    Alt: there is correlation

```python
telco_clean_df.total_charges.nunique()
```

```python
telco_clean_df['total_charges'].describe()
```

```python
telco_charges = telco_clean_df[telco_clean_df.total_charges != 0]
telco_charges.total_charges.describe()
```

```python
new_column = pd.qcut(telco_charges['total_charges'], 4, labels = ['a','b','c','d'])
new_column
```

```python
telco_charges['charge_rank']= new_column
```

```python
observed3 = pd.crosstab(telco_charges.charge_rank, telco_clean_df.churn)
observed3
```

```python
chi2, p, degf, expected = stats.chi2_contingency(observed3)
chi2, p, degf, expected
```

#### p is less than alpha so the alt hypothesis is supported
